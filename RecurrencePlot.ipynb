```python
sigma = 16.0
rho = 45.0
beta = 4.0
state0 = [10.0, -2.34199, 53.8658]
t = np.arange(0.0, 2.0, 0.001)

def Lorenz(fun,t):
    x,y,z=fun
    derivative=[sigma * (y - x), x * (rho - z) - y, x * y - beta * z]
    return derivative
    

states = odeint(Lorenz, state0, t)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(states[:,0], states[:,1], states[:,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.xlim(-40,40)
plt.ylim(-40,40)
plt.show()
```
